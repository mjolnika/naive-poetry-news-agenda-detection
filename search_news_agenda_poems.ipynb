```python
with open('topics2013.tsv', encoding='utf-8') as f:
  topicsfile = f.read().split('\n')

topicsdict = {}
for topic in topicsfile:
  if len(topic.split('\t')) == 2:
    topic, nums = topic.split('\t')
    topicsdict[topic] = nums.split(',')
```

```python
with open('chain_filter2013-3.tsv', encoding='utf-8') as f2:
  lines = f2.readlines()
```

```python
def showpoems(yourtheme):
  numbers = topicsdict[yourtheme]
  for line in lines:
    for num in numbers:
      if line.startswith(f'poem#{num}-'):
        id = line.split('\t')[0]
        date = '-'.join(line.split('\t')[1:4])
        tags = line.split('\t')[4]
        text = line.split('\t')[5]
        print('\n'.join([id, date, tags, text]))
```

```python
yourtheme = 'Встреча Милонова со Стивеном Фраем'
showpoems(yourtheme)
```
