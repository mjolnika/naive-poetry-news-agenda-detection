```python
import os
import re
import csv
import json
from collections import defaultdict, Counter
import datetime, time
```

```python
year = 2013
dct2013 = defaultdict(list)
with open(f'rises_u{year}.tsv', encoding='utf-8') as f:
    for line in f.readlines():
        token = line.split('\t')[0]
        periods = line.strip('\n').split('\t')[1:]
        yeared = []
        for period in periods:
            yeared.append(period)
        dct2013[token] = yeared


with open(f'rises_b{year}.tsv', encoding='utf-8') as f:
    for line in f.readlines():
        token = line.split('\t')[0]
        periods = line.strip('\n').split('\t')[1:]
        yeared = []
        for period in periods:
            yeared.append(period)
        dct2013[token] = yeared
 
        
dct2013 = dict(dct2013)
```

```python
def findnewsandpoems(token, periods, lag=10):  # расскомментируйте, если нужно напечатать новости тоже
    dates = []                                 # обычно их выводится в разы больше
    for period in periods:
        for i in range(lag):
            new_date = datetime.datetime.strptime(period, '%Y-%m-%d') + datetime.timedelta(i)
            string_date = new_date.strftime('%Y\t%m\t%d')
            dates.append(string_date)
    dates = sorted(dates)
    
    news_num = []
    with open('2013lentavzglyad.tsv', encoding='utf-8') as lv:
        datareader = csv.reader(lv, delimiter='\t')
        for num, row in enumerate(datareader):
            line_date = f'{row[0]}\t{row[1]}\t{row[2]}'
            if line_date in dates:
                if re.search(f'(?:^| |\.)({token})(?:$| |\.)', row[3]) is not None:
                    news_num.append(num)
                    
    with open(f'2013newsraw.tsv', encoding='utf-8') as t:
        datareader = csv.reader(t, delimiter='\t')
        for num, row in enumerate(datareader):
            if num in news_num:
                #print(row)
                if row[-1] == 'lenta':
                    print(row[0][:300])
                else:
                    print(row[:3], row[3])
                
    months = set([dat.split('\t')[1] for dat in dates])
    
    
    for month in months:
        print('*'*100)
        line_nums = []
        with open(f'preprocessed/mystem-2013-{month}.tsv', encoding='utf-8') as t:
            datareader = csv.reader(t, delimiter='\t')
            for num, row in enumerate(datareader):
                line_date = f'{row[0]}\t{row[1]}\t{row[2]}'
                if line_date in dates:
                    if re.search(f'(?:^| |\.)({token})(?:$| |\.)', row[3]) is not None:
                        line_nums.append(num)
                       # print(row[3], num, month)
                        
        with open(f'filtered/filtered-2013-{month}.txt', encoding='utf-8') as t:
            datareader = t.readlines()
            for line_num in line_nums:
                print(datareader[line_num])
```
