```python
import csv, json
import datetime

with open(f'periods-both{year}.tsv', encoding='utf-8') as f:
    period_dict = {}
    for line in f.readlines():
        period, _ = line.split('\t', 1)
        key = period.replace('-','\t')
        dates = []
        for i in range(10):
            new_date = datetime.datetime.strptime(period, '%Y-%m-%d') + datetime.timedelta(i)
            string_date = new_date.strftime('%Y\t%m\t%d')
            dates.append(string_date)
        period_dict[key] = dates
        
date_dict = {}
for key, value in period_dict.items():
    for date in value:
        date_dict[date] = key
```

```python
year = 2013
#poemfile = f'ffound_poems{year}.tsv'
poemfile = r'ffound_poems2013.tsv'
newsraw = f'{year}raw.tsv'
newslemmatized = f'{year}lentavzglyad.tsv'
written = 0
with open(f'{year}fcomparison.tsv', 'w', encoding='utf-8') as com:
    with open(poemfile, encoding='utf-8') as f:
        with open(newslemmatized, encoding='utf-8') as n:
            with open(newsraw, encoding='utf-8') as r:
                datareader = f.readlines()
                newsd = n.readlines()
                rawreader = r.readlines()
                
                for num, line in enumerate(datareader):
                    line = line.split('\t')
                    date = '\t'.join(line[:3])
                    tags = json.loads(line[3].replace("'", "\""))
                    #print(tags)
                    text = '\t'.join(line)
                    
                    period = date_dict[date]
                    dates = period_dict[period]
                    
                    
                    news_to_write = []
                    for newsnum, newsline in enumerate(newsd):
                        newsdate = '\t'.join(newsline.split('\t')[:3])
                        #print(newsdate, dates)
                        if newsdate in dates:
                            found_tags = []
                            
                            tokens = set(newsline.split('\t')[3].strip(' ').split(' '))
                            for tag in tags:
                                if tag in tokens:
                                    found_tags.append(tag)
                                    
                            if len(found_tags)>1:
                                
                                print('found', found_tags)
                                newstext = rawreader[newsnum].split('\t')[-1]
                                #print(newstext)
                                news_to_write.append(f'news\t{newsdate}\t{found_tags}\t{newstext}')
                    if news_to_write:
                        written += 1
                        com.write(f'poem#{written}\t{text}')
                        for n in news_to_write:
                            com.write(n)
```
